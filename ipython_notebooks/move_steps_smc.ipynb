```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N_particles = 10000
dim = 10
posterior_mean = np.ones(dim)*2
parameters = {'dim':dim, 'N_particles':N_particles, 'posterior_mean':posterior_mean, 'std':np.sqrt(2.)}

def log_prior(x, parameters=parameters):
    return multivariate_normal.logpdf(x, mean=np.zeros(parameters['dim']), cov=parameters['std']**2)
def log_target(x, parameters=parameters):
    return multivariate_normal.logpdf(x, mean=parameters['posterior_mean'])
def grad_log_target(x, parameters=parameters):
    return x-np.atleast_2d(parameters['posterior_mean'])
```

```python
def rw_kernel(particles, log_target, parameters):
    noise = np.random.normal(size=particles.shape)
    proposal = particles+noise
    diff_log_accept = log_target(proposal, parameters=parameters)-log_target(particles, parameters=parameters)
    u_log = np.log(np.random.random(size=particles.shape[0]))
    accepted = u_log<diff_log_accept
    new_particles = np.zeros(particles.shape)
    new_particles[accepted,:] = proposal[accepted,:]
    new_particles[~accepted,:] = particles[~accepted,:]
    acceptance_rate = accepted.mean()
    return new_particles, acceptance_rate

def resampling_step(particles, weights):
    if np.abs(weights.sum()-1.)>0.00001:
        weights = weights/weights.sum()
    cumulative_sum = np.cumsum(weights)
    cumulative_sum[-1] = 1.  # avoid round-off errors: ensures sum is exactly one
    indices =  np.searchsorted(cumulative_sum, np.random.random(len(weights)))
    return particles[indices,:]
```

```python
M = 500
particles_array = np.zeros((N_particles, dim, M))
# importance_sampling_step
particles = np.random.normal(size=(N_particles, dim))*parameters['std']
particles_array[:,:,0] = particles
log_weights = log_target(particles)-log_prior(particles)
weights = np.exp(log_weights)
particles = resampling_step(particles, weights)
particles_array[:,:,1] = particles

if True:
    list_squared_gradients = []
    list_mean_particles = []
    list_acceptance_rate = []
    grad_particles = grad_log_target(particles, parameters=parameters)
    list_squared_gradients.append(np.sum(grad_particles.mean(axis=0)**2))
    list_mean_particles.append (np.sum(particles.mean(axis=0)))
    for k in range(2,M):
        particles, acceptance_rate = rw_kernel(particles, log_target, parameters)
        particles_array[:,:,k] = particles
        grad_particles = grad_log_target(particles, parameters=parameters)
        list_squared_gradients.append(np.sum(grad_particles.mean(axis=0)**2))
        list_mean_particles.append (np.sum(particles.mean(axis=0)))
        list_acceptance_rate.append(acceptance_rate)
```

```python
m = 4
plt.scatter(x=particles_array[:,0,0], y=particles_array[:,1,0], color="b", alpha=0.2)
plt.scatter(x=particles_array[:,0,m], y=particles_array[:,1,m], color="r", alpha=0.2); plt.show()
```

```python
plt.plot(list_squared_gradients); plt.title("Error gradient"); plt.yscale('log'); plt.show()
plt.plot(list_mean_particles); plt.title("Error mean"); plt.show()
plt.plot(list_acceptance_rate); plt.title("acceptance rate"); plt.show()
```
