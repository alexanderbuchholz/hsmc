```python
import numpy as np
df = 5.
dim = 2
N_particles = 100000
res1 = np.random.normal(size=(N_particles, dim))
res2 = np.random.chisquare(df, size=(N_particles,1))
res = res1/np.sqrt(res2/df)
```

```python
res_wrong = np.random.standard_t(df, size=(N_particles, dim))
```

```python
res.var(axis=0), res_wrong.var(axis=0)
```

```python
import pandas as pd
frame1 = pd.DataFrame({"1" : res[:,0], "2" : res[:,1]})
frame1['type'] = 'true'
frame2 = pd.DataFrame({"1" : res_wrong[:,0], "2" : res_wrong[:,1]})
frame2['type'] = 'false'
frame_all = pd.concat([frame1, frame2])
#frame_all
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lmplot(x='1', y='2', hue='type', data=frame_all, fit_reg=False)
```

```python
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

# Draw the two density plots
ax = sns.kdeplot(res[:,0], res[:,1],
                 cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(res_wrong[:,0], res_wrong[:,1],
                 cmap="Blues", shade=True, shade_lowest=False)
```

```python
sns.jointplot(res[:,0], res[:,1], kind='kde')
sns.jointplot(res_wrong[:,0], res_wrong[:,1], kind='kde')
```
